```python
import pandas as pd
```

```python
df = pd.read_csv('atividade-prestacao-sns-24-para-a-sindrome-gripal.csv', sep=';')
```

```python
df.head()
```

```python
df = df.rename(columns = {
    'Dia da Semana': 'dia_semana',
    'Valor Absoluto' : 'valor_absoluto',
    'Taxa' : 'taxa',
    'Indicador' : 'indicador',
    'Número da Semana': 'nr_semana',
     df.columns[1] : 'periodo' #weird encoding
} )

df = df.drop(columns=['RowID'])
```

```python
df.loc[df.indicador == 'Encaminhamentos p/ Outros', 'indicador'] = 'encam_outros'
df.loc[df.indicador == 'Encaminhamentos p/ Autocuidados', 'indicador'] = 'encam_autocuidados'
df.loc[df.indicador == 'Encaminhamentos p/ Serviço Urgência', 'indicador'] = 'encam_urgencia'
df.loc[df.indicador == 'Encaminhamentos p/ CIAV', 'indicador'] = 'encam_CIAV'
df.loc[df.indicador == 'Encaminhamentos p/ INEM', 'indicador'] = 'encam_INEM'
df.loc[df.indicador == 'Encaminhamentos p/ CSP', 'indicador'] = 'encam_CSP'
```

```python
df = df.sort_values(by=['periodo', 'indicador'])
```

```python
df.head()
```

```python
#https://stackoverflow.com/questions/22798934/pandas-long-to-wide-reshape-by-two-variables

df2 = df.pivot(index='periodo',columns= 'indicador',
        values = ['valor_absoluto', 'taxa'] ).reset_index()
```

```python
df2.head()
```

```python
df2.columns = ['_'.join(col) if col[0]!='periodo' else col[0] for col in df2.columns]
```
