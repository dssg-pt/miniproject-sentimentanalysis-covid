```python
import pandas as pd
```

```python
df = pd.read_csv('atendimentos-nos-csp-gripe-orig.csv', sep=';')
```

```python
df = df.rename(columns = {
    'Nº Consultas nos CSP': 'n_cons_csp',
    'Nº Consultas CSP (Não Programadas)' : 'n_cons_csp_np',
    'Nº Consultas CSP (Programadas)' : 'n_cons_csp_p',
    'Nº Consultas Gripe nos CSP' : 'n_cons_csp_gripe'
} )
```

```python
df.loc[df.ARS == 'ARS Alentejo', 'ARS'] = 'alentejo'
df.loc[df.ARS == 'ARS Algarve', 'ARS'] = 'algarve'
df.loc[df.ARS == 'ARS Centro', 'ARS'] = 'centro'
df.loc[df.ARS == 'ARS Lisboa e Vale do Tejo', 'ARS'] = 'lisboa'
df.loc[df.ARS == 'ARS Norte', 'ARS'] = 'norte'
```

```python
df = df.sort_values(by=['Período', 'ARS'])
```

```python
df.head()
```

```python
#https://stackoverflow.com/questions/22798934/pandas-long-to-wide-reshape-by-two-variables

df2 = df.pivot(index='Período',columns= 'ARS',
        values = ['n_cons_csp_gripe', 'n_cons_csp_p', 'n_cons_csp_np', 'n_cons_csp_gripe'] ).reset_index()
```

```python
df2.head()
```

```python
df2.columns = ['_'.join(col) if col[0]!='Período' else col[0] for col in df2.columns]
```
